```python
import matplotlib.pyplot as plt
import numpy as np

# Simuler des donnes de montisation (revenus alatoires)
# Crons un tableau avec 1000 valeurs alatoires reprsentant des revenus
revenus = np.random.normal(loc=100, scale=20, size=1000)  # Moyenne de 100 et cart-type de 20

# Crer un histogramme des revenus
plt.figure(figsize=(10, 5))  # Taille de la figure
plt.subplot(1, 2, 1)  # Crer un sous-graphique (1 ligne, 2 colonnes, premier graphique)
plt.hist(revenus, bins=30, edgecolor='black', color='skyblue')
plt.title('Histogramme de Montisation')
plt.xlabel('Revenus')
plt.ylabel('Frquence')

# Crer un graphique de progression des revenus (graphe en ligne)
# Ici, nous voulons tracer les revenus au fil du temps, donc on prend l'indice comme axe X
plt.subplot(1, 2, 2)  # Deuxime sous-graphique
plt.plot(revenus, marker='o', color='green', linestyle='-', markersize=3)
plt.title('Progression des Revenus')
plt.xlabel('Temps (ou Indice)')
plt.ylabel('Revenus')

# Ajuster l'espacement entre les graphiques
plt.tight_layout()

# Afficher les graphiques
plt.show()
```
